ANALYSING AGE AND MARITAL STATUS

```python
import pandas as pd
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (10,5), title = 'Product Distribution by Age')
```

```python
for i in df['Age'].unique():
    print(i, ":",len(df[df['Age'] == i]))
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])

data = pd.DataFrame(lst, columns = ['Age','Products'])
data
```

```python
data.plot.bar(x = 'Age')
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind = 'bar', figsize = (10,5), title = 'Amount spend by age')
```

```python
df.groupby('Age').mean()['Purchase']
```

```python
df.groupby('Age').mean()['Purchase'].plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('Marital_Status').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('Gender').size().plot(kind = 'pie', autopct = '%.1f')
```
