ANALYSING MULTIPLE COLUMNS

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (8,4), title = 'Product Distribution by Age')
```

```python
sns.set(rc = {'figure.figsize' : (10,5)})
sns.countplot(x = "Age", hue = 'Gender', data = df)
```

```python
sns.set(rc = {'figure.figsize' : (10,5)})
sns.countplot(x = "Gender", hue = 'Age', data = df)
```

```python
sns.countplot(x = df['City_Category'])
```

```python
sns.countplot(x = 'Age', hue = 'City_Category', data = df)
```

```python
sns.countplot(x = 'City_Category', hue = 'Age', data = df)
```

```python
sns.countplot(x = 'City_Category', hue = 'Gender', data = df)
```

```python
df.groupby('City_Category').sum()['Purchase'].plot(kind = 'pie', autopct = "%0.1f")
```
